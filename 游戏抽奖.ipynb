```python
import pyautogui
from PIL import ImageGrab, Image
import numpy as np
from time import sleep
import numpy as np

import matplotlib.pyplot as plt
def check_color_on_screen(left, top, width, height):
    """
    检查屏幕指定区域是否存在红色或黄色
    :param left: 指定区域左上角的 x 坐标
    :param top: 指定区域左上角的 y 坐标
    :param width: 指定区域的宽度
    :param height: 指定区域的高度
    :return: 如果存在红色或黄色返回 True，否则返回 False
    """
    # 捕获屏幕指定区域的图像
    screenshot = ImageGrab.grab(bbox=(left, top, left + width, top + height))
    # 将图像转换为 numpy 数组
    img_array = np.array(screenshot)
    # 检查红色和黄色是否存在
    # 红色的 RGB 范围（简单示例，可按需调整）
    red_mask = (img_array[:, :, 0] > 220) & (img_array[:, :, 1] < 50) & (img_array[:, :, 2] < 50)
    # 黄色的 RGB 范围（简单示例，可按需调整）
    yellow_mask = (img_array[:, :, 0] > 200) & (img_array[:, :, 1] > 200) & (img_array[:, :, 2] < 100)
    # 判断是否存在红色或黄色
    has_red_or_yellow = np.any(red_mask) or np.any(yellow_mask)
    return has_red_or_yellow,screenshot

def click_action_a():
    # 定义点击行为 A
    print("点击最大")
    # 这里可以添加具体的点击操作，例如点击屏幕上的某个位置
    pyautogui.click(100, 100)

def click_action_b():
    # 定义点击行为 B
    print("执行点击行为 B")
    # 这里可以添加具体的点击操作，例如点击屏幕上的另一个位置
    pyautogui.click(200, 200)
```

```python
def click_max():
    # 定义点击行为 A
    print("点击最大",end='->')
    # 这里可以添加具体的点击操作，例如点击屏幕上的某个位置
    pyautogui.click(819, 1232)
```

```python
def click_coin():
    # 定义点击行为 A
    print("点击抽奖",end='->')
    # 这里可以添加具体的点击操作，例如点击屏幕上的某个位置
    pyautogui.click(631, 1331)
```

```python
def click_abandan():
    # 定义点击行为 A
    print("点击放弃")
    # 这里可以添加具体的点击操作，例如点击屏幕上的某个位置
    pyautogui.click(604, 1423)
```

```python
def sleep_rand():
    sleep(np.random.rand()+0.5)
```

```python
def get_pixel_color(x, y):
    """
    获取屏幕上指定位置的像素颜色
    :param x: 像素的 x 坐标
    :param y: 像素的 y 坐标
    :return: 像素的 RGB 值
    """
    screenshot = ImageGrab.grab()
    pixel_color = screenshot.getpixel((x, y))
    return pixel_color

# 指定要获取颜色的位置
x, y = 308, 656
color = get_pixel_color(x, y)
print(f"屏幕坐标 ({x}, {y}) 的颜色为 RGB: {color}")
```
